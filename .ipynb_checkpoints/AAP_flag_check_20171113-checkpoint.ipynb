```python
import pandas as pd
```

```python
f1 = 'G:\\マイドライブ\\02_MDM\\work\\#201712\\1213\\mdm_aap_20171213.txt'
```

```python
f2 = 'G:\\マイドライブ\\02_MDM\\work\\#201712\\1213\\plm_aap_20171213.xlsx'
```

```python
mdm = pd.read_csv(f1, sep='\t', dtype='object', engine='python')
```

```python
plm = pd.read_excel(f2, header=1, dtype='object')
```

```python
mdm.shape
```

```python
plm.shape
```

```python
merge = pd.merge(mdm, plm, left_on=['Item code', 'ColorID'], right_on=['item_trading_code', 'color_code'], how='outer', indicator=True)
```

```python
merge.iloc[:10]
```

```python
add = merge[merge['_merge'] == 'right_only']
```

```python
add.shape
```

```python
add.iloc[:10]
```

```python
add.columns
```

```python
add1 = add[['item_trading_code','color_code', 'mp_region']]
```

```python
add1.iloc[:10]
```

```python
add1.loc[:,'Product classification'] = 1
```

```python
add1.iloc[:10]
```

```python
add1.rename(columns={'item_trading_code':'Item code', 'color_code':'ColorID', 'mp_region':'Target System'}, inplace=True)
add1.iloc[:10]
```
